```python
directory = '/home/yoyo/Downloads/kaggle_amazon_jpg/planet/planet/'
label_csv_file = directory + 'train_classes.csv'
img_dir = directory + 'train-jpg/'
output_dir = 'data/satellite/amazon_new/'
```

```python
import glob
import os
from PIL import Image
import numpy as np
from shutil import copyfile
import pandas as pd

# np.random.seed(42)
```

```python
df = pd.read_csv(label_csv_file)
df['agri'] = df['tags'].apply(lambda x: 'agriculture' in x)
```

```python
df
```

```python
idx1 = df['agri'] == 1
agri = df[idx1]
agri = agri.sample(frac=0.8)
agri
```

```python
idx0 = df['agri'] == 0
noagri = df[idx0]
noagri = noagri.sample(frac=0.4)
noagri
```

```python
n = len(agri)
n_train = int(n*0.05)
n_test = int(n*0.1)
train_agri = agri.iloc[:n_train]
test_agri = agri.iloc[n_train:n_test]
unlabel_agri = agri.iloc[n_test:]

n = len(noagri)
n_train = int(n*0.05)
n_test = int(n*0.1)
train_noagri = noagri.iloc[:n_train]
test_noagri = noagri.iloc[n_train:n_test]
unlabel_noagri = noagri.iloc[n_test:]
```

```python
def write_to_dir(d, image_dir, out_dir):
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)
    count = 0
    for i, row in d.iterrows():
        image_name = row['image_name'] + '.jpg'
        src = image_dir + image_name
        dst = out_dir + image_name
        if os.path.exists(src) and not os.path.exists(dst):
            copyfile(src, dst)
            count += 1
    print(count)
```

```python
write_to_dir(train_agri, img_dir, output_dir + 'train/agri/')
write_to_dir(test_agri, img_dir, output_dir + 'test/agri/')
write_to_dir(unlabel_agri, img_dir, output_dir + 'unlabeled/agri/')

write_to_dir(train_noagri, img_dir, output_dir + 'train/noagri/')
write_to_dir(test_noagri, img_dir, output_dir + 'test/noagri/')
write_to_dir(unlabel_noagri, img_dir, output_dir + 'unlabeled/noagri/')
```
