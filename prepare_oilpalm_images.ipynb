```python
directory = '/home/yoyo/Downloads/widsdatathon2019/'
label_csv_file = directory + 'traininglabels.csv'
img_dir = directory + 'train_images/train_images/'
output_dir = 'data/satellite/oilpalm_extra/'
```

```python
import glob
import os
from PIL import Image
import numpy as np
from shutil import copyfile
import pandas as pd

np.random.seed(42)
```

```python
df = pd.read_csv(label_csv_file)
```

```python
df
```

```python
idx1 = df['has_oilpalm'] == 1
palm = df[idx1]
palm = palm.sample(frac=1)
palm
```

```python
idx0 = np.random.choice(np.where(df['has_oilpalm'] != 1)[0], 1000, replace=False)
nopalm = df.iloc[idx0]
nopalm
```

```python
n = len(palm)
n_train = int(n*0.1)
n_test = int(n*0.2)
train_palm = palm.iloc[:n_train]
test_palm = palm.iloc[n_train:n_test]
unlabel_palm = palm.iloc[n_test:]

n = len(nopalm)
n_train = int(n*0.1)
n_test = int(n*0.2)
train_nopalm = nopalm.iloc[:n_train]
test_nopalm = nopalm.iloc[n_train:n_test]
unlabel_nopalm = nopalm.iloc[n_test:]
```

```python
def write_to_dir(d, image_dir, out_dir):
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)
    count = 0
    for i, row in d.iterrows():
        image_name = row['image_id'][:9] + row['image_id'][13:]
        src = image_dir + image_name
        dst = out_dir + image_name
        if os.path.exists(src) and not os.path.exists(dst):
            copyfile(src, dst)
            count += 1
    print(count)
```

```python
write_to_dir(train_palm, img_dir, output_dir + 'train/palm/')
write_to_dir(test_palm, img_dir, output_dir + 'test/palm/')
write_to_dir(unlabel_palm, img_dir, output_dir + 'unlabeled/palm/')

write_to_dir(train_nopalm, img_dir, output_dir + 'train/nopalm/')
write_to_dir(test_nopalm, img_dir, output_dir + 'test/nopalm/')
write_to_dir(unlabel_nopalm, img_dir, output_dir + 'unlabeled/nopalm/')
```
